```python
import pandas as pd
import numpy as np
```

```python
notas_df = pd.read_excel('Data\\Notas.xlsx')
```

```python
notas_df['Promedio'] = notas_df.mean(axis=1)
notas_df
```

```python
notas_df['Aprobado/Desaprobado'] = np.where(notas_df['Promedio'] >= 10.5, 'Aprobado', 'Desaprobado')
```

```python
notas_df
```

```python
condiciones = [
    (notas_df['Promedio'] >= 18),
    (notas_df['Promedio'] < 18) & (notas_df['Promedio'] >= 11),
    (notas_df['Promedio'] < 11)
]
```

```python
resultados = ['A', 'B', 'C']
```

```python
notas_df['Rango'] = np.select(condiciones, resultados)
```
